```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
x = np.linspace(-10,10,200)
plt.plot(x,sigmoid(x))
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

â€¢ Make a line plot of the difference between the weekly maximum high price and the weekly minimum low price for Facebook. This should be a single line.

```python
path = 'C:/Users/khoat/OneDrive - National Economics University/Desktop/1. Study/3.1. Data Preparation & Visualiazation/data/'
fb = pd.read_csv(path + 'fb.csv')
fb['week'] = pd.DatetimeIndex(fb.date).week
fb.head()
```

â€¢ Plot the bar chart displaying the number of earthquakes and tsunamis in Indonesia each day.

```python
earthquake['date'] = pd.to_datetime(earthquake.time, unit='ms')
earthquake
```

```python
df = earthquake.groupby(pd.DatetimeIndex(earthquake.date).date).agg({'mag':'count', 'tsunami':'sum'})
df.rename(columns={'mag':'earthquake'},inplace=True)
```

```python
df.plot.bar(figsize=[19,4])
```
